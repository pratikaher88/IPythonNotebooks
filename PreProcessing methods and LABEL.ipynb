```python
from sklearn import preprocessing
import numpy as np

data = np.array([[2.2, 5.9, -1.8,1.1], [5.4, -3.2, -5.1,1.1], [-1.9, 4.2, 3.2,1.1]])
```

```python
binData= preprocessing.Binarizer(threshold=1.5).transform(data)
```

```python
print(binData)
```

```python
mean_data= data.mean(axis=0)
std= data.std(axis=0)
print(mean_data,std)
```

```python
scaled_data=preprocessing.scale(data)
```

```python
print(scaled_data.mean(axis=0),scaled_data.std(axis=0))
```

SCALING

```python
minmax= preprocessing.MinMaxScaler(feature_range=(0,0.5))
data_minmax= minmax.fit_transform(data)
print(data_minmax)
```

Normalization

```python
dataL1= preprocessing.normalize(data,norm='l1')
dataL2= preprocessing.normalize(data,norm='l2')
print(dataL1)
print(dataL2)
```

```python
import pandas as pd
from IPython.display import display

data = pd.read_csv('smsspam.txt', header=None, index_col=False, names=['age', 'workclass', 'fnlwgt', 'education', 
                                                                      'education-num', 'marital-status', 'occupation', 
                                                                      'relationship', 'race', 'gender', 'capital-gain', 
                                                                      'capital-loss', 'hours-per-week', 'native-country', 
                                                                      'income'])
```

```python
data.head()
```

```python
print(data.columns)
dummies=pd.get_dummies(data)
print(dummies.columns)
```

```python
data = data[['age', 'workclass', 'education', 'gender', 'hours-per-week', 'occupation', 'income']]
display(data)
```

```python
print(data.columns)
dummies=pd.get_dummies(data)
print(dummies.columns)
```

```python
features= dummies.ix[:,'age':'occupation_ Transport-moving']
X=features.values
y=dummies['income_ >50K'].values
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test= train_test_split(X,y,random_state=0)
logreg=LogisticRegression()
logreg.fit(X_train,y_train)
```

```python
print(logreg.score(X_test,y_test))
```
